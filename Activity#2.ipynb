```python
def traverse_array(arr):
     """Function to traverse and print all elements in the array"""
     print("Traversing the array:")
     for i, element in enumerate(arr):
        print(f"Index {i}: {element}")

def reverse_array(arr):
     """Function to reverse and print all elements in the array"""
     print("Reversing the array:")
     reversed_arr = arr[::-1]
     print(reversed_arr)

def sort_array(arr):
     """Function to sort and print all elements in the array"""
     print("Sorting the array:")
     arr.sort()
     print(arr)

def insert_array(arr):
     """Function to insert and print all elements in the array"""
     index = int(input("Enter the index to insert at:"))
     value = int(input("Enter the value to insert:"))
     arr.insert(index, value)
     print("After the insertion:", arr)

def delete_array(arr):
     """Function to delete and print all elements in the array"""
     number = int(input("Enter a number to delete:"))
     if number in arr:
        arr.remove(number)
        print("After the deletion:", arr)
     else:
        print(f"{number} not found in the array")

def modify_array(arr):
    """Function to modify an element in the array by index"""
    print("Current array:", arr)
    index = int(input("Enter the index to modify(0-4): "))

    if 0 <= index < len(arr):
        print(f"Current value at index {index} is {arr[index]}")
        new_value = int(input("Enter the new value: "))
        arr[index] = new_value
        print("Array after modification:", arr)
    else:
        print(f"Invalid index. Please enter a number between 0 and {len(arr) - 1}.")

def main():
    #initialize an array
    arr = [10, 20, 30, 40, 50]
    print("Original array:", arr)

    while True:
      print("\nMenu:")
      print("1. Traverse array")
      print("2. Reverse array")
      print("3. Sort array")
      print("4. Insert array")
      print("5. Delete array")
      print("6. Modify array")
      print("7. Exit")
      choice = input("Enter your choice (1-7): ")

      if choice == '1':
          traverse_array(arr)
      elif choice == '2':
          reverse_array(arr)
      elif choice == '3':
          sort_array(arr)
      elif choice == '4':
          insert_array(arr)
      elif choice == '5':
          delete_array(arr)
      elif choice == '6':
          modify_array(arr)
      elif choice == '7':
          break
      else:
          print("Invalid choice. Pick a number (1-7)")


if __name__ == "__main__":
  main()
```
